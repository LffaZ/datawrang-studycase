# Study Case 2: Akses dari Excel (3 Sheets)

```python
file_path = 'National_Data_Appendix.xlsx'

# Membaca semua sheet dalam file Excel
xls = pd.ExcelFile(file_path)

# Menampilkan nama sheet yang ada
print(xls.sheet_names)

# Membaca setiap sheet berdasarkan nama sheet
sheet1 = pd.read_excel(xls, sheet_name='Veteran')
sheet2 = pd.read_excel(xls, sheet_name='Non-Veteran')
sheet3 = pd.read_excel(xls, sheet_name='Recent Veteran VHA User')

sheet1.columns = sheet1.columns.str.replace('\n', ' ')
sheet2.columns = sheet2.columns.str.replace('\n', ' ')
sheet3.columns = sheet3.columns.str.replace('\n', ' ')

sheet1.columns = sheet1.columns.str.replace('Veteran ', '')
sheet1.head()
```

```python
sheet2.columns = sheet2.columns.str.replace('Non-Veteran ', '')
sheet2.head()
```

```python
sheet3.columns = sheet3.columns.str.replace('VHA Veteran ', '')
sheet3.head()
```

```python
df_combined = pd.concat([sheet1, sheet2, sheet3], ignore_index=True)
df_combined.head()
```
